```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
CityData = pd.read_csv(city_data_to_load)
RideData = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merged_data = pd.merge(RideData, CityData,how = 'left', on=["city", "city"])

# Display the data table for preview
merged_data.head()
```

```python
#City Types
urbanData = merged_data[merged_data["type"] == "Urban"]
suburbanData = merged_data[merged_data["type"] == "Suburban"]
ruralData = merged_data[merged_data["type"] == "Rural"]

#Calculations and plots

urbanRideCount = urbanData.groupby(["city"]).count()["ride_id"]
urbanAvg = urbanData.groupby(["city"]).mean()["fare"]
urbanDriverCount = urbanData.groupby(["city"]).mean()["driver_count"]

suburbRideCount = suburbanData.groupby(["city"]).count()["ride_id"]
suburbAvg = suburbanData.groupby(["city"]).mean()["fare"]
suburbDriverCount = suburbanData.groupby(["city"]).mean()["driver_count"]

ruralRideCount = ruralData.groupby(["city"]).count()["ride_id"]
ruralAvg = ruralData.groupby(["city"]).mean()["fare"]
ruralDriverCount = ruralData.groupby(["city"]).mean()["driver_count"]
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(urbanRideCount, urbanAvg, s=5*urbanDriverCount, c="red", edgecolor="black", linewidths=1, label="Urban")

plt.scatter(suburbRideCount, suburbAvg, s=5*suburbDriverCount, c="purple", edgecolor="black", linewidths=1, label="Suburban")

plt.scatter(ruralRideCount, ruralAvg, s=5*ruralDriverCount, c="yellow",edgecolor="black", linewidths=1, label="Rural")

# Incorporate the other graph properties
plt.title("Ride Sharing Data")
plt.ylabel("Average Fare ")
plt.xlabel("Total Number of Rides Per City")
plt.grid(True)

# Create a legend
legend = plt.legend(fontsize="small", 
                  numpoints=1,  title="City Types", 
                  labelspacing=0.5)
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(45, 40, "Circle size correlates with driver count per city.")

# Save Figure
plt.savefig('Images')
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
Types_fare_sum = merged_data.groupby(["type"]).sum()["fare"]
Fare_sum= merged_data['fare'].sum()
Fare_Percent_perTypes = 100 * Types_fare_sum / Fare_sum
# Build Pie Chart
plt.pie(Fare_Percent_perTypes, labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], explode=[0, .1, .1],
        autopct='%1.1f%%',shadow=True, startangle=90)
# Save Figure
plt.savefig('Images')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ridePercents = 100 * merged_data.groupby(["type"]).count()["ride_id"] / merged_data["ride_id"].count()

# Build Pie Chart
plt.pie(ridePercents, labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], autopct='%1.1f%%', explode=[0, .1, .1],shadow=True, startangle=90)
plt.title("% of Total Rides (City Type)")

# Save Figure
plt.savefig("Images")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driverPercents = 100 * merged_data.groupby(["type"]).sum()["driver_count"] / merged_data["driver_count"].sum()

# Build Pie Charts
plt.pie(driverPercents, labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], explode=[.3, .2, .2], 
        autopct='%1.1f%%', shadow=True, startangle= 70)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Images")

# Show Figure
plt.show()
```
