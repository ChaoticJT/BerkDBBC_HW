```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
  
# players and their purchase counts
players_purchases = purchase_data["SN"].value_counts()
CountPlayer_purchase = players_purchases.count()
purchase_data
```

## Player Count

* Display the total number of players


```python
ID_DF = purchase_data.loc[: , ["Age" , "Gender" , "SN"]]
ID_DF = ID_DF.drop_duplicates()
PlayerCount = ID_DF.count()[0]
print(f"The total number of active players : {PlayerCount}")
print(f"The total number of players who purchased an item : {CountPlayer_purchase}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of Unique Items
itemDF = purchase_data["Item Name"]
UniqueItemCount = itemDF.value_counts()
itemTotal = UniqueItemCount.count()

# Average Purchase Price
avg_price = purchase_data["Price"].mean()

# Total Number of Purchases
TotalPurch = ID_DF.count()

# Total Revenue
income = purchase_data["Price"]
Revenue = income.sum()

print(f"the total number of purchases : {TotalPurch} , the Total Revenue : {Revenue}")
print(f"the number of items : {itemTotal} , the average price : {avg_price}")

statistics = {
    "Analysis" : ["# of items" , "Avg Price" , "Total Purchases" , "Total Revenue"],
    "Values" : [itemTotal , avg_price , TotalPurch , Revenue]
}
DF = pd.DataFrame(statistics, columns=["Analysis" , "Values"])
DF
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
genderAVG = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
genderCnt = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
genderDF = pd.DataFrame({"Purchase Count": genderCnt, 
                         "Average Purchase Price": genderAVG, 
                         "Total Purchase Value": gender_purchase_total})
genderDF
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 12 , 18, 21, 24, 29, 34, 39, 99]
group_names = ["<13", "10-18", "19-20", "21-24", "25-29", "30-34", "35-39", "40+"]

ID_DF["Age Ranges"] = pd.cut(ID_DF["Age"], age_bins, labels=group_names)

age_totals = ID_DF["Age Ranges"].value_counts()
age_percentage = age_totals/ PlayerCount* 100
age_demographics = pd.DataFrame({"Total Count": age_totals, "Percentage of Players": age_percentage})
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
age_average = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

age_data = pd.DataFrame({"Purchase Count": age_counts, 
                         "Average Purchase Price": age_average, 
                         "Total Purchase Value": age_purchase_total})
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
UsersTotal_pur = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
TopUser_averagePur = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
TopUser_countPur = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")

user_data = pd.DataFrame({"Total Purchase Value": UsersTotal_pur, 
                          "Average Purchase Price": TopUser_averagePur, 
                          "Purchase Count":TopUser_countPur})

user_data_sorted = user_data.sort_values("Total Purchase Value", ascending=False)
user_data_sorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
ItemData = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

TotalPurchaseValue = ItemData.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
average_item_purchase = ItemData.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = ItemData.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

ItemData_DF = pd.DataFrame({"Total Purchase Value": TotalPurchaseValue, "Item Price": average_item_purchase, "Purchase Count": item_count})

ItemData_DF = ItemData_DF.sort_values("Purchase Count", ascending=False)
ItemData_DF
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Item Table (Sorted by Total Purchase Value)
TopProfitItems = ItemData_DF.sort_values("Total Purchase Value", ascending=False)

TopProfitItems.head(5)
```
